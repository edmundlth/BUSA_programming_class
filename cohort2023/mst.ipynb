Write a function called `triplet` that takes in a possibly empty string as input and returns
the set of most frequently occurring character triplets in the string. 

A character triplet is three adjacent characters occurring as a substring of the input.
The string "abcabc" has the following triplet substrings: "abc", "bca", "cab", "abc"
in order from left to right. The most frequent triplet in that string is "abc", because it
occurs twice, and all of the other triplets occur once.


Note that the output of the triplet func.on should always be a Python set. For the
purposes of this ques.on it does not maGer which order the items in the set are displayed
by Python.


Strings with length less than three do not contain any triplets. In such cases the result should
be the empty set.



Examples:
```
>>> triplet("abcabc")
{'abc'}
>>> triplet("abracadabra")
{'abr', 'bra'}
>>> triplet("abcabcabbabcd")
{'abc'}
>>> triplet("aaaa")
{'aaa'}
>>> triplet("ab")
set()
```

````python
def make_triplet_tally(string):
    # maps all triplets to their frequency
    rec = {}
    if len(string) < 3:
        return rec
    for i in range(len(string) - 2):
        trigram = string[i: i + 3]
        if trigram not in rec:
            rec[trigram] = 0
        rec[trigram] += 1
    return rec


test_inputs = [
    "abcabc", 
    "aaaa", 
    "ab", 
    "", 
]

test_outputs = [
    {"abc": 2, "bca": 1, "cab": 1}, 
    {"aaa": 2}, 
    dict(), 
    dict()
]

# run unit test
for inpt, output in zip(test_inputs, test_outputs):
    print(make_triplet_tally(inpt) == output)
````

````python
def triplet(string):
    # make a dictionary recording the frequency of all triplets in `string`
    tally = make_triplet_tally(string)
    
    # Figure out what is the highest frequency triplet. 
    highest_freq = max(tally.values())
    
    # Output all triples with that frequency as a set!
    result = set()
    for trigram, freq in tally.items():
        if freq == highest_freq:
            result.add(trigram)
    return result
````

````python
ALLOWED_OPERATOR = ["+", "-"]

def check_legal(expr):
    parts = expr.split()
    
    if len(parts) == 1:
        if not parts[0].isdigit():
            return None
    
    for p in parts:
        if not p.isdigit() and p not in ALLOWED_OPERATOR:
            return None
    return 

def plus_op(result, current_num):
    return result + current_num

def minus_op(result, current_num):
    return result - current_num

OPERATOR = {
    "+": plus_op, 
    "-": minus_op
}

def caculate(expr):
    result = 0
    parts = expr.split() # a list of strings of either integer or ["+", "-"]
    current_op = "+"
    expect_num = True
    
    for p in parts:
        if expect_num:
            if not p.isdigit():
                return None
            current_num = int(p)
            result = OPERATOR[current_op](result, current_num)
            expect_num = False
        else:
            if p not in ALLOWED_OPERATOR:
                return None
            current_op = p
            expect_num = True
    
    if expect_num: 
        return None
    return result

print(caculate("1 + 1 + 3 + "))
````
