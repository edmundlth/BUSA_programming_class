Write a function which takes a filename as a string and opens the file, printing statistics for every line in the file: the line number, amount of words, amount of characters. Remember to close the file when you’re done!

```python
def line_stats(line):
    num_words = len(line.split())
    num_char = len(line)
    return num_words, num_char


def file_stats(file_name):
    with open(file_name) as file:
        line_num = 1
        for line in file:
            num_words, num_char = line_stats(line)
            print(f"Line num: {line_num}, Num words: {num_words}, Num character: {num_char}")
            line_num += 1
    return


file_stats("./testinput_small.fastq")
```

Write a function which takes a string as input and returns a sorted list of the words which occur only once in the string.

```python
def make_tally(words):
    """
    Takes a list of words and return a dictionary containing their word counts
    """
    tally = dict()
    for w in words:
        if w not in tally:
            tally[w] = 1
        else: 
            tally[w] += 1
    return tally

def sorted_unique_words(text):
    tally = make_tally(text.split())
    result = []
    for word in tally:
        if tally[word] == 1:
            result.append(word)
    return sorted(result)

def sorted_unique_words_using_count(text):
    words = text.split()
    unique_words = set(words)
    result = []
    for w in words:
        if words.count(w) == 1 and w not in result:
            result.append(w)
    return sorted(result)

test_text = "Hello world! hello hello, a, a a a world, bye."
words = test_text.split()
sorted_unique_words(test_text)
```

```python
def get_words_from_file(filename):
    file = open(filename)
    content = file.read()
    file.close()
    return content.split()


def get_word_freq(wordlist):
    word_freq ={}
    for word in wordlist:
        if word not in word_freq:
            word_freq[word] = 1
        else:
            word_freq[word] += 1
    return word_freq


def get_highest_freq(freqs):
    return max(freqs)

def get_word_with_freq(word_freq_dict, freq):
    words = []
    for key in word_freq_dict:
        if word_freq_dict[key] == freq:
            words.append(key)
    return words

def approximate_song(filename):
    # parse the file, i.e. get a list of words
    words = get_words_from_file(filename)
    
    # Obtain word frequency 
    # get a dictionary word_freq of type dict(string -> int)
    word_freq = get_word_freq(words)

    # Obtain highest frequency
    highest_frequency = get_highest_freq(word_freq.values())
    
    # Obtain words with highest frequency
    # type(highest_freq_words) = list(str)
    highest_freq_words = get_word_with_freq(word_freq, highest_frequency)
    
    # ordered words with highest freq
    ordered_words = sorted(highest_freq_words)
    
    # return the first 
    
    return ordered_words[0]
```
