```python
circles = [
     {'col':'orange','xy':(100,100)}, 
     {'col':'blue',  'xy':(125,150)},
     {'col':'blue',  'xy':( 75,150)},
     {'col':'blue',  'xy':(125, 50)},
     {'col':'blue',  'xy':( 75, 50)},
     {'col':'blue',  'xy':(150,100)},
     {'col':'blue',  'xy':( 50,100)}
]

ORANGE_RADIUS = 10

def BLUE_RADIUS(i): return (30 - i % 20)

for i in range(100):
    print('''plot(0,0,type='n', xlab="", ylab="", axes=FALSE, xlim=c(0,200), ylim=c(0,200))''')
    for c in circles:
        # draw at circle['xy'] in color circle['col']
        if c['col'] == 'blue':
            radius = BLUE_RADIUS(i)
        else:
            radius = ORANGE_RADIUS
        print("symbols({}, {}, circle={}, inch=FALSE, add=TRUE, fg=NA, bg='{}')".format(c['xy'][0], c['xy'][1], radius, c['col']))
```
